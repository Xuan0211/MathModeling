```python
import numpy as np
import pandas as pd
df = pd.read_excel("../data/附件3.xlsx")
df['日期'] = pd.to_datetime(df['日期'])
df = df.set_index('日期')
df['单品编码'] = df['单品编码'].astype(str)
print(df)
```

```python
df_after = pd.DataFrame([])
df_after['销售日期']=pd.date_range(start='2020-07-01', end='2023-06-30', freq='D')
df_after = df_after.set_index('销售日期')
for index, row in df.iterrows():
    df_after.at[index, row['单品编码']] = row['批发价格(元/千克)']
df_after = df_after.fillna(0)
print(df_after)
df_after.to_csv('../data/成本价按日期单品合并.csv',sep=',',encoding='utf_8_sig')
```
