```python
import numpy as np
import pandas as pd
price = pd.read_csv("../data/销售额按日期单品合并.csv",encoding='utf-8',index_col='销售日期')
volume = pd.read_csv("../data/销售量按日期单品合并.csv",encoding='utf-8',index_col='销售日期')
pl = pd.read_excel("../data/附件1.xlsx")
pl['单品编码'] = pl['单品编码'].astype(str)
pl = pl.set_index('单品编码')
```

```python
print(price)
print(volume)
```

```python
types_per_price = pd.DataFrame([])
for types in ['花叶类', '花菜类', '水生根茎类', '茄类', '辣椒类', '食用菌']:
    type_price = pd.DataFrame([])
    type_volume = pd.DataFrame([])
    for index in price:
        plrow = pl.loc[index]
        if plrow['分类名称'] == types:
            name_CN = plrow['单品名称']
            type_price[name_CN] = price[index]
            type_volume[name_CN] = volume[index]
    #print(type_price)
    #print(type_volume)
    type_price['Sum'] = type_price.apply(lambda row: row.sum(), axis=1)
    type_volume['Sum'] = type_volume.apply(lambda row: row.sum(), axis=1)
    type_per_price = type_price['Sum']/type_volume['Sum']
    types_per_price[types] = type_per_price
    #print(type_per_price)
```

```python
print(types_per_price)
types_per_price.to_csv('../data/各品类每日单价.csv',sep=',',encoding='utf_8_sig')
```
