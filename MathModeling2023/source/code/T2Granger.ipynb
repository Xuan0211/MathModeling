```python
import numpy as np
import pandas as pd
emotion = pd.read_csv("../data/各品类每日代理指数.csv",encoding='utf-8',index_col='销售日期')
price = pd.read_csv("../data/各品类每日单价.csv",encoding='utf-8',index_col='销售日期')
```

```python
from statsmodels.tsa.stattools import grangercausalitytests
for types in ['花叶类', '花菜类', '水生根茎类', '茄类', '辣椒类', '食用菌']:
    print('-'*40)
    print(types+': 代理指数 with 单价')
    temp = pd.DataFrame([])
    temp['emotion'] = emotion[types]
    temp['price'] = price[types].values
    grangercausalitytests(temp[['emotion','price']], maxlag=8)
    print('-'*40)
    print(types+': 单价 with 代理指数')
    grangercausalitytests(temp[['price','emotion']], maxlag=8)
```

```python
volume = pd.read_csv("../data/销售量按日期品类合并.csv",encoding='utf-8',index_col='销售日期')
```

```python
from statsmodels.tsa.stattools import grangercausalitytests
for types in ['花叶类', '花菜类', '水生根茎类', '茄类', '辣椒类', '食用菌']:
    print('-'*40)
    print(types+': 销售量 with 单价')
    temp = pd.DataFrame([])
    temp['volume'] = volume[types]
    temp['price'] = price[types].values
    grangercausalitytests(temp[['volume','price']], maxlag=8)
    print('-'*40)
    print(types+': 单价 with 销售量')
    grangercausalitytests(temp[['price','volume']], maxlag=8)
```
